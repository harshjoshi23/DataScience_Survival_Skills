## Loading Dataset :

## Plotting Data : 

### For Regression 1 & 2 : 

```python
plt.figure(figsize=(15,5))
# for Regression 1
plt.subplot(1, 3, 1)
plt.scatter(regression1.iloc[:, 0], regression1.iloc[:, 1])
plt.title('Regression 1 Dataset')
plt.xlabel('X')
plt.ylabel('Y')

# for Regression 2
plt.subplot(1, 3, 2)
plt.scatter(regression2.iloc[:, 0], regression2.iloc[:, 1])
plt.title('Regression 2 Dataset')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```

#### Comments on Graphs : 
<b>Regression 1 (regression1):</b> The data appears to be linear with some variance around the regression line. It could represent a scenario where there is a direct relationship between the independent and dependent variables with some noise.<br><br>
<b>Regression 2 (regression2):</b> This dataset has a non-linear distribution. A linear model may not be the best fit for this data as it won't capture the curvature present in the data points. A polynomial or a non-linear regression model might be more appropriate.<br><br>


### For Classification : 

```python
plt.figure(figsize=(15,5))
plt.subplot(1, 3, 3)
plt.scatter(classification.iloc[:, 0], classification.iloc[:, 1], c=classification.iloc[:, 2])
plt.title('Classification Dataset')
plt.xlabel('Feature1')
plt.ylabel('Feature2')

plt.tight_layout()
plt.show()
```

#### Comment on classification model : 
<b>Classification:</b> The classification dataset shows two groups that are somewhat linearly separable. The decision boundary from a linear SVM seems to separate the two classes reasonably well, but there might be some misclassified points close to the boundary.
