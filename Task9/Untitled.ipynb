```python
# Task 1: Code Optimization Example

# Original function for demonstration (replace with your actual task)
def sum_squares(n):
    total = 0
    for i in range(n):
        total += i ** 2
    return total

# Optimized function using list comprehension
def optimized_sum_squares(n):
    return sum([i ** 2 for i in range(n)])

# Task 2: Pi Approximation Function
def approximate_pi(N):
    count_inside = 0
    for _ in range(N):
        x, y = np.random.rand(2)
        distance = x**2 + y**2
        if distance <= 1.0:
            count_inside += 1
    return 4 * count_inside / N

# Parallelization of Pi Approximation
def parallel_pi_calculation(N_values):
    with Pool(processes=4) as pool:  # Adjust number of processes based on your machine
        results = pool.map(approximate_pi, N_values)
    return results

# Task 3: Speed Optimization with Numba
@jit(nopython=True)
def numba_approximate_pi(N):
    count_inside = 0
    for _ in range(N):
        x, y = np.random.rand(2)
        distance = x**2 + y**2
        if distance <= 1.0:
            count_inside += 1
    return 4 * count_inside / N

# Task 4: Result Presentation
def plot_results(N, pi_estimates):
    plt.plot(N, pi_estimates, label='Estimated Pi')
    plt.axhline(y=np.pi, color='r', linestyle='-', label='Actual Pi')
    plt.xlabel('Number of Points')
    plt.ylabel('Pi Approximation')
    plt.legend()
    plt.show()

N_values = [1000, 10000, 100000, 1000000]  # Example N values for Pi approximation
start_time = time.time()
pi_estimates = [approximate_pi(N) for N in N_values]  # Sequential execution
print(f"Sequential execution time: {time.time() - start_time}")

start_time = time.time()
pi_estimates_parallel = parallel_pi_calculation(N_values)  # Parallel execution
print(f"Parallel execution time: {time.time() - start_time}")

start_time = time.time()
pi_estimates_numba = [numba_approximate_pi(N) for N in N_values]  # Numba optimized execution
print(f"Numba optimized execution time: {time.time() - start_time}")

plot_results(N_values, pi_estimates_numba)  # Plotting the results
```
