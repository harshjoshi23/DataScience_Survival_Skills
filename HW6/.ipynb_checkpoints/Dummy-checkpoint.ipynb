```python
import h5py
import numpy as np
import matplotlib.pyplot as plt



filename = 'rectanglesdsss.sec'
with h5py.File(filename, 'r') as file:
    ground_truth_group = file['/ground_truth']
    predicted_group = file['/predicted']

    ground_truth_rectangles = []
    predicted_rectangles = []

    for key in ground_truth_group.keys():
        # # Access the subgroups inside the 'ground_truth' and 'predicted' groups
        # print(f"{key}th ground_truth_group: {ground_truth_group}")
        # print(f"{key}th predicted_group: {predicted_group}")

        # Access the attributes of each subgroup directly
        gt_subgroup = ground_truth_group[key]
        pred_subgroup = predicted_group[key]

        if 'i0' in gt_subgroup.attrs and 'i0' in pred_subgroup.attrs:
            gt_attr = gt_subgroup.attrs['i0']
            pred_attr = pred_subgroup.attrs['i0']

            # Check if the attributes are single float values
            if isinstance(gt_attr, (float, np.float64)):
                gt_rect = (gt_attr,)
            else:
                gt_rect = tuple(gt_attr)

            if isinstance(pred_attr, (float, np.float64)):
                pred_rect = (pred_attr,)
            else:
                pred_rect = tuple(pred_attr)

            ground_truth_rectangles.append(gt_rect)
            predicted_rectangles.append(pred_rect)


def calculate_iou(rect1, rect2):
    x1, y1, w1, h1 = rect1
    x2, y2, w2, h2 = rect2

    intersection_x = max(0, min(x1 + w1, x2 + w2) - max(x1, x2))
    intersection_y = max(0, min(y1 + h1, y2 + h2) - max(y1, y2))

    intersection_area = intersection_x * intersection_y
    union_area = w1 * h1 + w2 * h2 - intersection_area
    print(f"{i}th has: {union_area} as Union area")
    iou = intersection_area / union_area
    return iou


iou_scores = [calculate_iou(gt_rect, pred_rect) for gt_rect, pred_rect in zip(ground_truth_rectangles, predicted_rectangles)]
print(iou_scores)

# Print the first few rectangles for verification
print("Ground Truth Rectangles:")
for i in range(min(5, len(ground_truth_rectangles))):
    print(f"Rect {i}: {ground_truth_rectangles[i]}")

print("\nPredicted Rectangles:")
for i in range(min(5, len(predicted_rectangles))):
    print(f"Rect {i}: {predicted_rectangles[i]}")
```

```python
# Calculate IoU scores

# Plot histogram
plt.hist(iou_scores, bins=20, edgecolor='black')
plt.xlabel('IoU Score')
plt.ylabel('Frequency')
plt.title('IoU Score Histogram')
plt.show()
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt



filename = 'rectanglesdsss.sec'
with h5py.File(filename, 'r') as file:
    ground_truth_group = file['/ground_truth']
    predicted_group = file['/predicted']

    ground_truth_rectangles = []
    predicted_rectangles = []

    for key in ground_truth_group.keys():
        # # Access the subgroups inside the 'ground_truth' and 'predicted' groups
        # print(f"{key}th ground_truth_group: {ground_truth_group}")
        # print(f"{key}th predicted_group: {predicted_group}")

        # Access the attributes of each subgroup directly
        gt_subgroup = ground_truth_group[key]
        pred_subgroup = predicted_group[key]

        if 'i0' in gt_subgroup.attrs and 'i0' in pred_subgroup.attrs:
            gt_attr = gt_subgroup.attrs['i0']
            pred_attr = pred_subgroup.attrs['i0']

            # Check if the attributes are single float values
            if isinstance(gt_attr, (float, np.float64)):
                gt_rect = (gt_attr,)
            else:
                gt_rect = tuple(gt_attr)

            if isinstance(pred_attr, (float, np.float64)):
                pred_rect = (pred_attr,)
            else:
                pred_rect = tuple(pred_attr)

            ground_truth_rectangles.append(gt_rect)
            predicted_rectangles.append(pred_rect)


def calculate_iou(rect1, rect2):
    if len(rect1) == 1:
        x1, y1, w1, h1 = rect1[0], 0, 0, 0
    else:
        x1, y1, w1, h1 = rect1

    if len(rect2) == 1:
        x2, y2, w2, h2 = rect2[0], 0, 0, 0
    else:
        x2, y2, w2, h2 = rect2

    intersection_x = max(0, min(x1 + w1, x2 + w2) - max(x1, x2))
    intersection_y = max(0, min(y1 + h1, y2 + h2) - max(y1, y2))

    intersection_area = intersection_x * intersection_y
    union_area = w1 * h1 + w2 * h2 - intersection_area

    if union_area == 0:
        iou = 0  # or handle it in a way that makes sense for your application
    else:
        iou = intersection_area / union_area

    return iou




iou_scores = [calculate_iou(gt_rect, pred_rect) for gt_rect, pred_rect in zip(ground_truth_rectangles, predicted_rectangles)]
print(iou_scores)

# Print the first few rectangles for verification
print("Ground Truth Rectangles:")
for i in range(min(5, len(ground_truth_rectangles))):
    print(f"Rect {i}: {ground_truth_rectangles[i]}")

print("\nPredicted Rectangles:")
for i in range(min(5, len(predicted_rectangles))):
    print(f"Rect {i}: {predicted_rectangles[i]}")
```

```python
# Calculate IoU scores

# Plot histogram
plt.hist(iou_scores, bins=20, edgecolor='black')
plt.xlabel('IoU Score')
plt.ylabel('Frequency')
plt.title('IoU Score Histogram')
plt.show()
```
