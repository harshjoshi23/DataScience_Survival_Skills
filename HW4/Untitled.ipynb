```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('./census_income_dataset.csv')
```

```python
df.head(5)
```

```python
df['AGE'].info()
```

### Age distribution of resspondets

```python
plt.figure(figsize=(10,6))
```

```python
#Age distribution of respondents (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df['AGE'], bins=20, kde=False, color='skyblue')
plt.title('Age Distribution of Respondents', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.savefig('age_distribution.svg')
plt.show()
```

```python
# b) Relationship status distribution (Bar chart)
plt.figure(figsize=(10, 6))
sns.countplot(df['RELATIONSHIP'], palette='pastel')
plt.title('Distribution of Relationship Status', fontsize=16)
plt.xlabel('Relationship Status', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.savefig('relationship_status_distribution.svg')
plt.show()
```

```python
# b) Relationship status distribution (Pie chart)
plt.figure(figsize=(10, 6))
relationship_counts = df['RELATIONSHIP'].value_counts()
plt.pie(relationship_counts, labels=relationship_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=90)
plt.title('Distribution of Relationship Status', fontsize=16)
plt.savefig('relationship_status_distribution_pie.svg')
plt.show()
```

```python
# c) Salary distribution within each educational level (Bar chart)
plt.figure(figsize=(12, 8))
education_salary = df.groupby(['EDUCATION', 'SALARY']).size().unstack()
education_salary.plot(kind='bar', stacked=True, colormap='coolwarm')
plt.title('Salary Distribution Within Each Educational Level', fontsize=16)
plt.xlabel('Education Level', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Salary', bbox_to_anchor=(1, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.savefig('salary_education_distribution.svg')
plt.show()
```

```python
# Enhance the plot
plt.figure(figsize=(12, 6))
sns.histplot(df['AGE'], bins=30, kde=True, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Respondents', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('age_distribution_updated.svg')
plt.show()
```

```python
# Enhance the plot
plt.figure(figsize=(10, 6))
relationship_counts = df['RELATIONSHIP'].value_counts()
explode = (0.1, 0, 0, 0, 0, 0)  # Emphasize the first slice
plt.pie(relationship_counts, labels=relationship_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=90, explode=explode)
plt.title('Distribution of Relationship Status', fontsize=16)
plt.savefig('relationship_status_distribution_pie_updated.svg')
plt.show()
```

```python
# Enhance the plot
plt.figure(figsize=(14, 8))
education_salary = df.groupby(['EDUCATION', 'SALARY']).size().unstack()
education_salary.plot(kind='bar', stacked=True, colormap='coolwarm', edgecolor='black')
plt.title('Salary Distribution Within Each Educational Level', fontsize=16)
plt.xlabel('Education Level', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Salary', bbox_to_anchor=(1, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.savefig('salary_education_distribution_updated.svg')
plt.show()
```
