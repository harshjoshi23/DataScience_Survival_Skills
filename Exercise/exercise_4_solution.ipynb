## Explore programming languages

```python
# Lets make a copy for us to work on
df = data_df.copy()
len(df)
```

```python
# Check if values are missing (NaN)
df["Your main technology / programming language"].isnull().sum()
```

```python
# Show rows with missing values
df[df["Your main technology / programming language"].isnull()]
```

```python
# Remove missing values
df = df[~df["Your main technology / programming language"].isnull()]
len(df)
```

```python
df["Your main technology / programming language"]
```

**$\rightarrow$ categorical/ nominal variable**

```python
# List all programming languages
df["Your main technology / programming language"].unique()
```

```python
# Convert to lowercase
df["Your main technology / programming language"] = df["Your main technology / programming language"].str.lower()
```

```python
# Top 10
top10 = df["Your main technology / programming language"].value_counts()[:10]
top10
```

```python
languages = "|".join(l for l in top10.index[:10])
languages
```

```python
# extract strings using regular expressions
pattern = r"(javascript|java|python|php|c\+\+|c#|swift|scala|go|kotlin)"
df["Your main technology / programming language"] = df["Your main technology / programming language"].str.extract(pattern)
```

```python
top10 = df["Your main technology / programming language"].value_counts()[:10]
top10
```

```python
counts = top10.sort_values()
categories = counts.index
count = counts.values
```

```python
plt.bar(x=categories, height=count)
plt.show()
```

```python
plt.barh(categories, count)
plt.show()
```

### Preparing a plot for a (scientific) figure
**Size and font size**
* Page width: about 7 inches (= 180mm, typical DinA4 page with borders)
* We want to fit two plots next to each other $\rightarrow$ about 3 inches each
* We set the font size to 8
* The font should be editable in Inkscape!

**Improving visuals**
* Declutter = Remove unnecessary elements such as plot borders, grid lines etc.
* Draw attention to the important elements
* Avoid 3D effects
* Use color sparingly, avoid rainbow colormaps etc.

```python
# Set font size
plt.rcParams['font.size'] = 12 # for now: 12, later: 8

# Make the text editable (in the SVG file)
plt.rcParams['svg.fonttype'] = 'none'
```

```python
plt.figure(figsize=(2.5, 3))  # 2.5 to 3 inches width
plt.title("Top 10 programming languages")
bars = plt.barh(categories, count)

# Add count next to each bar
plt.bar_label(bars, padding=5)

# Remove chart border & y-axis (declutter!)
plt.xlim([0, 250])
sns.despine(left=True)
plt.tick_params(left=False)

plt.xlabel('Number of respondents')

# Save as PNG and SVG
plt.savefig("plots/barplot_languages.png", bbox_inches="tight", dpi=150)
plt.savefig("plots/barplot_languages.svg", bbox_inches="tight")
plt.show()
```

```python
sns.histplot(df, x="Age", bins=bins, alpha=0.5)

# Compute average age
mean_age = df["Age"].mean()
print("Mean:", mean_age)

# Compute median
median_age = df["Age"].median()
print("Median:", median_age)

# Compute mode
mode_age = df["Age"].mode().values[0]
print("Mode:", mode_age)

# Highlight mean, median, mode by plotting a vertical line
plt.axvline(x=mean_age, ls='--', lw=2, color='tab:orange', label=f'Mean: {mean_age:.0f} years')
plt.axvline(x=median_age, ls='--', lw=2, color='tab:green', label=f'Median: {median_age:.0f} years')
plt.axvline(x=mode_age, ls='--', lw=2, color='tab:red', label=f'Mode: {mode_age:.0f} years')
plt.legend()
plt.show()
```

```python
# Width: 3 inches
plt.figure(figsize=(3, 2))
plt.title("Age distribution")

# Histogram
sns.histplot(df, x='Age', bins=bins, alpha=0.5)

# Highlight mean, median, mode by plotting a vertical line
plt.axvline(x=mean_age, ls='--', lw=1, color='tab:orange', label=f'Mean: {mean_age:.0f} years')
plt.axvline(x=median_age, ls='--', lw=1, color='tab:green', label=f'Median: {median_age:.0f} years')
plt.axvline(x=mode_age, ls='--', lw=1, color='tab:red', label=f'Mode: {mode_age:.0f} years')

# Declutter: Remove the upper and right border
plt.ylim([0,200])
sns.despine(trim=False)
plt.legend(frameon=False)

# Save as SVG
plt.savefig("plots/age_histogram.svg", bbox_inches="tight")

plt.show()
```

## Exploring age and total years of experience

```python
# Let's make a copy
df = data_df.copy()
len(df)
```

```python
# remove NaN's from "Total years of experience"
df = df[~df["Total years of experience"].isnull()]
len(df)
```

```python
plt.figure(figsize=(12, 5))
plt.scatter(df["Total years of experience"], df["Age"])
_ = plt.xticks(rotation=90)
```

```python
# Check datatype of the column
df["Total years of experience"].dtype
```

Pandas datatype **object (O)**: String or mixed numeric and non-numeric values

```python
# replace non-numeric values in column "Total years of experience" with NaN
df["Total years of experience"] = df["Total years of experience"].apply(pd.to_numeric, errors='coerce')

# remove NaN's from "Total years of experience" again
df = df[~df["Total years of experience"].isnull()]
len(df)
```

```python
# Check datatype of the column
df["Total years of experience"].dtype
```

```python
# plot it again
plt.figure(figsize=(12, 5))
plt.scatter(df["Total years of experience"], df["Age"])
```

**Remove outliers**

```python
df["Total years of experience"].plot.hist(bins=50)
plt.xlabel("Total years of experience")
plt.xlim(0,100)
plt.show()
# --> positively skewed distribution

# boxplot_pd = df["Total years of experience"].plot.box(showfliers=False, return_type='dict')#
boxplot = plt.boxplot(df["Total years of experience"], showfliers=False)
plt.ylabel("Years")
plt.show()
```

```python
[item.get_ydata() for item in boxplot['whiskers']]
```

```python
df['Total years of experience'].describe()
```

```python
# Q3 + whis*(Q3-Q1)
upper_limit = 12 + 1.5 * (12 - 5)
upper_limit
```

```python
# Q1 - whis*(Q3-Q1)
lower_limit = 5 - 1.5 * (12 - 5)
lower_limit
```

```python
# Apply the limits to our DataFrame
df = df[(df["Total years of experience"] <= upper_limit) & (df["Total years of experience"] >= lower_limit)]
```

```python
# plot it again without outliers
plt.scatter(df["Total years of experience"], df["Age"])
plt.xlabel("")
plt.show()
```

**Overplotting:** Points overlap and obscure each other $\rightarrow$ smaller size, transparency (alpha)

```python
plt.title("Age vs. years of experience")
sns.scatterplot(data=df, x="Total years of experience", y="Age", alpha=0.3, s=10)
plt.xlabel("Years of experience")
plt.ylabel("Age in years")
sns.despine()

# Save as SVG
plt.savefig("plots/age_experience.svg", bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(3, 2))
plt.title("Age vs. years of experience")
plt.scatter(df["Total years of experience"], df["Age"], s=2, alpha=0.6, edgecolors="none")
plt.xlabel("Years of experience")
plt.ylabel("Age in years")
sns.despine()

# Save as SVG
plt.savefig("plots/age_experience.svg", bbox_inches="tight")
plt.show()
```

## Analyzing salary & seniority level
Now, let's explore the yearly brutto salary and how it differs for seniority levels. <br>
We will look at different plot types for visualizing distributions and for comparing distributions among groups.

```python
# Let's make a copy
df = data_df.copy()
```

```python
df = df.rename(columns={"Yearly brutto salary (without bonus and stocks) in EUR": "Yearly brutto salary in EUR"})
```

```python
# Boxplot of salary distribution
sns.boxplot(data=df, y='Yearly brutto salary in EUR')
```

```python
df['Yearly brutto salary in EUR'].describe()
```

```python
# Q3 + whis*(Q3 - Q1)
upper_limit = 80000 + 1.5*(80000 - 58800)

# Q1 - whis*(Q3 - Q1)
lower_limit = 58800 - 1.5*(80000 - 58800)

# Apply limits to DataFrame
df = df[(df["Yearly brutto salary in EUR"] <= upper_limit) &
        (df["Yearly brutto salary in EUR"] >= lower_limit)]
```

```python
sns.boxplot(data=df, y="Yearly brutto salary in EUR")
plt.axhline(y=df["Yearly brutto salary in EUR"].median(), color="red")
plt.show()
```

```python
sns.barplot(data=df, y="Yearly brutto salary in EUR" , ci="sd", capsize=0.2, errwidth=2)
plt.axhline(y=df["Yearly brutto salary in EUR"].mean(), color="red")
plt.show()
```

```python
# Remove NaN's from column "Seniority level"
df = df[~df['Seniority level'].isnull()]
```

```python
sns.boxplot(data=df, y="Yearly brutto salary in EUR", x="Seniority level")
plt.show()
```

```python
top5 = df["Seniority level"].value_counts()[:5].index
top5
```

```python
# Remove rows, only keep top 5 seniority levels
df = df[df["Seniority level"].isin(top5)]
```

```python
sns.boxplot(data=df, y="Yearly brutto salary in EUR", x="Seniority level")
plt.show()
```

```python
sns.violinplot(data=df, y="Yearly brutto salary in EUR", x="Seniority level")
```

```python
sns.barplot(data=df, y="Yearly brutto salary in EUR", x="Seniority level", ci="sd", capsize=0.2, errwidth=2)
plt.show()
```

```python
# Show raw values
sns.pointplot(data=df, y="Yearly brutto salary in EUR", x="Seniority level", ci="sd", linestyles="none", capsize=0.2)
plt.show()
```

```python
# Show raw values
sns.stripplot(data=df, y="Yearly brutto salary in EUR", x="Seniority level", s=3)
plt.show()
```

```python
sns.swarmplot(data=df, y="Yearly brutto salary in EUR", x="Seniority level")
plt.show()
```

```python
# Order seniority levels by their median salary
order = df.groupby("Seniority level")["Yearly brutto salary in EUR"].median().sort_values()
```

```python
order
```

```python
plt.figure(figsize=(3.5, 2))
plt.title("Salary by seniority level")

sns.boxplot(data=df, y="Yearly brutto salary in EUR", x="Seniority level", palette="pastel", linewidth=1, showfliers=False, order=order.index)
sns.stripplot(data=df, y="Yearly brutto salary in EUR", x="Seniority level", s=2, order=order.index)

sns.despine()

# Save as SVG
plt.savefig("plots/salary_seniority.svg", bbox_inches="tight")
plt.show()
```

## Colormaps

### Sequential colormaps


```python
sns.color_palette("Blues", as_cmap=True)
```

### Perceptually uniform sequential colormaps
= equal steps in values are perceived as equal steps in the color space <br>
Great for displaying heatmaps for example

```python
sns.color_palette("viridis", as_cmap=True)
```

```python
sns.color_palette("plasma", as_cmap=True)
```

```python
# Example: "Flights" dataset
flights = sns.load_dataset("flights")
sns.heatmap(flights.pivot(index="month", columns="year", values="passengers"), cmap="plasma")
plt.show()
```

### Diverging colormaps
Use these if there is a meaningful middle point. <br>
Example: Spearman correlation (-1: negative correlation, 0: no correlation, 1: positive correlation)

```python
sns.color_palette("RdBu", as_cmap=True)
```

```python
sns.color_palette("coolwarm", as_cmap=True)
```
